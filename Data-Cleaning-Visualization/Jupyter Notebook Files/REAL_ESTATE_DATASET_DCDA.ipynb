## Import required libraries


```python
import numpy as np
import pandas as pd 
```

## Read Dataset

```python
data=pd.read_csv('https://raw.githubusercontent.com/FrancescoMariottini/project3/main/inputs/all_sales_data.csv',sep=',',dtype='unicode')
```

## Original Data

```python
data.head(5)
```

## Assign selected columns to data_new variable

```python
data_new=data[['open_fire','area','land_surface','land_plot_surface','rooms_number']]
```

## Functions to process each column for cleaning

```python
def process_open_fire_col():# function to replace open_fire data with 1s and 0s
    dt = data_new['open_fire']
    dt.convert_dtypes()
    dt = dt.str.lower()
    dt = dt.map({'false': 0, 'true': 1, 'nan': -999, '0': -999})
    dt.fillna(-999, inplace=True)
    dt = dt.astype(int)
    return dt


def process_area_col():# function to replace area data with 1s and 0s
    data_area = data_new['area']
    data_area = pd.to_numeric(data_area, errors='coerce')
    data_area.fillna(-999, inplace=True)
    return data_area


def process_land_surf_col():# function to replace land_surface data with 1s and 0s
    data_land_surf = data_new['land_surface']
    data_land_surf = pd.to_numeric(data_land_surf, errors='coerce')
    data_land_surf.fillna(-999, inplace=True)
    return data_land_surf


def process_land_plot_col():# function to replace land_plot_surface data with 1s and 0s
    data_land_plot = data_new['land_plot_surface']
    data_land_plot = pd.to_numeric(data_land_plot, errors='coerce')
    data_land_plot.fillna(-999, inplace=True)
    return data_land_plot

def process_num_rooms_col():# function to replace room_numbers data with 1s and 0s
    data_rmo = data_new['rooms_number']
    data_rmo = pd.to_numeric(data_rmo, errors='coerce')
    data_rmo.fillna(-999, inplace=True)
    data_rmo=data_rmo.astype('Int64')
    return data_rmo
```

## Calling Functions

```python
data_new['open_fire']=process_open_fire_col()
data_new['area']=process_area_col()
data_new['land_surface']=process_land_surf_col()
data_new['land_plot_surface']=process_land_plot_col()
data_new['rooms_number']=process_num_rooms_col()
```
